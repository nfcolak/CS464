In this assignment, you will use a custom dataset, which you need a wrapper class to access its data. For this purpose, implement a custom dataset class to organize data with the specifications you need. This class will load batches of data using a PyTorch DataLoader later. The class signature of CustomDataset class is provided to you in the cell below, where you can implement your custom dataset.

**IMPORTANT:** As you may have difficulties while fitting all images to your RAM, you need to read the images while getting that item (each item is an image here). We encourage you to read the data files in the "__getitem__" function to avoid encountering any issues related to fitting the data to memory. This way, you do not need to hold the dataset in a single variable, which would be memory intensive.

```python
class CustomDataset(Dataset):
    
    def __init__(self, data, labels, **kwargs):
        self.data = data
        self.labels = labels

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        image = Image.open(self.data[index], "r")
        image = image.convert("RGB")
        image = image.resize((224,224))
        image = np.asarray(image,dtype = "float")
        
        def normalize(image):
            return (image - np.min(image)) / (np.max(image) - np.min(image))
        
        normalized_img = normalize(image)
        label = self.labels[index]
        return normalized_img, label
```

### Fully Connected Network (FCN) Model [8 points]
The second model you will try is a Fully Connected Network composed of Linear layers and non-linear activation functions connected to each other. For this part, the number of features for your linear layers is [1024, 512, 256]. Again, you can use ReLU as your non-linear activation function in your model (except the classification layer).

```python
class FCNet(nn.Module):
    '''Define your full connected network'''
    def __init__(self, **kwargs):
        super(FCNet, self).__init__()
        self.fc1 = nn.Linear(224*224*3, 1024) 
        self.fc2 = nn.Linear(1024, 512)
        self.fc3 = nn.Linear(512, 256)
        self.relu = nn.ReLU()
        self.out = nn.Linear(256, 4)

    def forward(self, x):
        x = torch.flatten(x,1) # flatten all dimensions except the batch dimension
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        x = self.relu(x)        
        x = self.out(x)
        return x
```

### Training [12 points]

```python
#HINT: note that your training time should not take many days.
#TODO:
#Pick your hyper parameters

max_epoch = 14
train_batch = 64
test_batch = 64
learning_rate = 0.1
momentum = 0.5
weight_decay = 5e-04

model_FCN = FCNet()
criterion = nn.CrossEntropyLoss()
min_vald_loss = np.inf
optimizer = torch.optim.SGD(model_FCN.parameters(), lr=learning_rate, weight_decay=5e-04) 

PATH_FCN_SGD = "/content/drive/MyDrive/CS464_HW3/best_fcn_sgd.pth"
min_vald_loss = np.inf
history = {'train_accuracy': [], 'vald_accuracy':[], 'train_loss': [], 'vald_loss': []}
    
for epoch in range(max_epoch):
    train_loss = 0.0
    train_correct = 0
    for i, data in enumerate(train_loader, 0):
        inputs, labels = data
        optimizer.zero_grad()
        input = inputs.permute(0, 3, 1, 2)
        outputs = model_FCN(input.float())
        preds = torch.argmax(outputs, dim=1)
        train_correct += torch.sum(preds==labels)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    with torch.no_grad():
        vald_loss = 0.0
        vald_correct = 0
        for data in validation_loader:
            images, labels = data
            images = images.permute(0, 3, 1, 2)
            outputs = model_FCN(images.float())
            preds = torch.argmax(outputs, dim=1)
            vald_correct += torch.sum(preds==labels)
            loss = criterion(outputs, labels)
            vald_loss += loss.item()

    print(f'Epoch {epoch+1}:\t\t train_loss: {(train_loss / len(train_loader)):.2f}\t\t train_accuracy: {(train_correct / (len(train_loader) * train_batch)):.2f}\t\t validation_loss: {(vald_loss / len(validation_loader)):.2f}\t\t validation_accuracy: {(vald_correct / (len(validation_loader) * test_batch)):.2f}')
    if (vald_loss / len(validation_loader)) < min_vald_loss:
        print(f'Saving new best FCN SGD with vald_loss: {(vald_loss / len(validation_loader)):.2f}')
        min_vald_loss = (vald_loss / len(validation_loader))
        torch.save(model_FCN.state_dict(), PATH_FCN_SGD)

    history['train_loss'].append(train_loss / len(train_loader))
    history['vald_loss'].append(vald_loss / len(validation_loader))
    history['train_accuracy'].append(train_correct / (len(train_loader) * train_batch))
    history['vald_accuracy'].append(vald_correct / (len(validation_loader) * test_batch))

print(f'Best validation loss: {min_vald_loss}')
```
